```python
import numpy as np
import pandas as pd
```

```python
#Intialize menu
menu = ['1. Add Employee Data', 
        '2. View Employee Data',
        '3. Update Employee Data',
        '4. Delete Employee Data',
        '5. Exit']

#Initialize data
data = {'Employee_id':np.arange(1,11),
            'Employee_name': ['Othuke', 'Frank', 'Samantha', 'Sancho', 'Marley', 
                              'Joseph', 'Ayobami', 'Blessing', 'Maguire', 'Tosin'],
            'Gender':['Male', 'Male', 'Female', 'Male', 'Male',
                     'Male', 'Female', 'Female', 'Male', 'Female'],
            'Role':['Manager', 'Product designer','HRM', 'Customer success analyst','Data Analyst',
                    'Compliance officer','Finance lead', 'Sales lead', 'Marketing Associate','Technical assistant'],
            'Salary':['500000', '150000', '170000', '200000', '250000', '180000','300000','250000','220000', '120000']}
df = pd.DataFrame(data)
```

```python
def create_employee_data():
    option = True
    
    while option !=2:
        print('\n Provide the following details to add a new employee record \n')
        employee_name = input('Enter Employee name: ')
        gender = input('Enter Employee Gender: ')
        role = input('Enter Employee Role: ')
        salary = int(input('Enter Employee Salary: '))
        df.loc[len(df)] = [len(df)+1, employee_name, gender, role, salary]
        option = int(input('\n Enter 1 to add another employee data or 2 to return to the main menu: '))
    return df 
```

```python
def read_features():
    option = True
    while option != 3:
        print('\n ========== Employee Data Report==============')
        print('1. View all Employee Data')
        print('2. View a particular Employee data')
        print('3. Return to main menu')
        option = int(input('\n Enter an option: '))
        if option == 1:
            print(df)
        elif option == 2:
            name = input('Enter Employee name: ')
            if name not in df['Employee_name'].tolist():
                print('This name is not in the Employee data collection')
            else:
                print(df[df['Employee_name'] == name])
        elif option == 3:
            break
```

```python
def update_features():
    option = True
    while option != 2:
        name = input('\n Enter the name of the Employee whose data you want to update: ')
        if name not in df['Employee_name'].values:
            print('\n This name is not in the data collection, try again\n ')
            break
        index = df[df['Employee_name'] == name].index
        for i, col in enumerate(df.columns):
            print(i+1, col)
        i = int(input('Enter the Serial number of the column you want to edit: '))
        if i not in range(len(df.columns)+1):
            print('This is not a valid column, please try again')
            break
        col = df.columns[i-1]
        data = input('Enter the updated data: ')
        df.loc[index, col] = data
        print('\n Data updated successfully')
        option = int(input('\n Enter 1 to update another feature or 2 to go back to the main menu: '))
        return df
```

```python
def delete_features():
    option = True
    while option != 2:
        name = input('\n Enter the name of the Employee whose data you want to delete: ')
        if name not in df['Employee_name'].values:
            print('\n This name is not in the data collection, try again\n ')
            break
        index = df[df['Employee_name'] == name ].index
        df.drop(index, inplace=True)
        print('\n Data deleted successfully')
        option = int(input('\n Enter 1 to delete another employe data or 2 to go back to the main menu: '))
        return df
```

```python
def Main_menu():
    option = True
    while option !=5:
        print('========Employee Data Records==========')
        for i in menu:
            print(i)
        option = int(input('\n Choose from the main menu (1-5): '))
        if option == 1:
            df = create_employee_data()
            print('\n Data added successfully \n')
        elif option == 2:
            read_features()
        elif option == 3:
            df = update_features()
        elif option == 4:
            df = delete_features()
        elif option == 5:
            print('\n Thank you for your time')
            break
        else:
            print('This is not a valid option, please try again\n ')

#Run the application
Main_menu()
        
```
